```python
import cv2
import numpy as np
import os

''' TODO
Cut frames in `bachelor1.mp4 (frame after the cut):
166
424
732
1052
1254
1478 - missed
1600 - missed
1698 - missed
1834
2078
2253
2431
2672 - missed
...

'''

# Constants
THRESHOLD_HSV_FRAME_COMPARISON = 15
video_path = './videos/bachelor1.mp4'

cap = cv2.VideoCapture(video_path)

frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
height = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
width = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
print(f"\nfilename: {os.path.basename(video_path)}\n \
      frames: {frames}\n \
      fps: {fps}\n \
      height: {height}\n \
      width: {width}\n")


#frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
prev_mean = 0
prev_frame = 0
jump_cut_frames = []
for frame_number in range(0, int(frames)):
    cap.set(1, frame_number-1)
    res, frame = cap.read()
    curr_mean = np.mean(frame, axis=(0,1,2))
    if abs(curr_mean - prev_mean) > THRESHOLD_HSV_FRAME_COMPARISON:    
        print('frame: {}, diff: {}, mean: {}'.format(frame_number, abs(curr_mean - prev_mean), curr_mean))
        cv2.imwrite('./jump_cut_images/'+str(frame_number)+'.png', frame)
        cv2.imwrite('./jump_cut_images/'+str(frame_number-1)+'.png', prev_frame)
        jump_cut_frames.append(frame_number)
    prev_mean = curr_mean
    prev_frame = frame
print(jump_cut_frames)
```

```python
### Numpy mean calculation notes

# shape is (2, 2, 3)
two_by_two_pixels = np.array([[[1, 2, 3], [3, 4, 5]], 
                              [[5, 6, 7], [7, 8, 9]]])

# for a 2d array, `mean` with `axis=0` calculates means along columns for each tuple value
# e.g. (avg 1,5; avg 2,6...etc.)
column_mean = np.mean(two_by_two_pixels, axis=(0)) # array([[3., 4., 5.], [5., 6., 7.]])
# for a 2d array, `mean` with `axis=0,1` calculates means by rows for each tuple, then by columns 
# e.g. (avg 1,3=2, then avg 5,7=6, finally avg 2,6=4)
rgb_mean = np.mean(two_by_two_pixels, axis=(0,1)) # array([4., 5., 6.])

# finally, using all 3 axis for a 2d array returns a single value
overall_mean = np.mean(two_by_two_pixels, axis=(0,1,2)) # 5.0
overall_mean
```
